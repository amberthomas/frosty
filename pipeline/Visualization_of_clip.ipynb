```python
import matplotlib.pyplot as plt
import numpy as np
import gdal

print('RGB visualization of the MODIS Surface Reflectance Product Image.')
# can use example image I gave or your own from your google drive
img = np.transpose(np.array(gdal.Open('Alaska_Anchorage_example.tif').ReadAsArray(), dtype='int16'),axes=(1,2,0))
print(img.shape)
print(img.shape[2]/7)
print(np.min(img[:,:,0:4]))
print(np.max(img[:,:,0:4]))
low_bound = (np.min(img[:,:,0:4])*-1)
high_bound = np.max(img[:,:,0:4])
img = img + np.ones(img.shape)*(low_bound)
div = (255*np.ones(img.shape))/(high_bound + low_bound)
sc_img = img*div
# band 1: red, band 2: NIR, band 3: blue, band 4: green
# stack order below should be RGB
rgb = np.dstack((sc_img[:,:,0],sc_img[:,:,3],sc_img[:,:,2]))
rgb8 = rgb.astype(np.uint8)
plt.figure(figsize=(10,10))
plt.imshow(rgb8)
plt.show()

"""
The smeared look of image is just due to resolution. To verify that the clip is correct you can go to
https://code.earthengine.google.com/ec8726b8a614725974c335cc2375735b
and compare the visualizations. 
"""

```
